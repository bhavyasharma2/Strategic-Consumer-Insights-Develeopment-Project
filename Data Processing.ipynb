So, from the above output it can be observed that there are 101 null values in 'product attributes' column and 1 null value in 'author' column and 1 null value in content column while all other columns have zero null values.

```python
#Replacing the null values in the dataframe with 'Not Available':

reviews_df['author'].fillna('Not available', inplace=True)
reviews_df['content'].fillna('Not available', inplace=True)
reviews_df['product_attributes'].fillna('Not available', inplace=True)

reviews_df.isna().sum()
```

```python
#Descriptive Statistics of the dataframe:
reviews_df.describe()
```

* Count - Gives the number of non-null values
* mean - The average of the values present in the column
* std - Standard deviation value for the data present in the column
* min - minimum value present in the given column
* max - maximum value present in the given column
* 25% - 25 percentile (the value at which 25% of the answers lie below that value and 75% of the answers lie above that value.)
* 50% - 50 percentile (Also known as the Median. The median cuts the data set in half.  Half of the answers lie below the median and half lie above the median.)
* 75% - 75 percentile (Also known as the third, or upper, quartile. The 75th percentile is the value at which 25% of the answers lie above that value and 75% of the answers lie below that value. Above the 75th or below the 25th percentile)

```python
#Calculating IQR for 'helpful_count':
Q1_helpful = reviews_df['helpful_count'].quantile(0.25)
Q3_helpful = reviews_df['helpful_count'].quantile(0.75)
IQR_helpful = Q3_helpful - Q1_helpful
lower_bound_helpful = Q1_helpful - 1.5 * IQR_helpful
upper_bound_helpful = Q3_helpful + 1.5 * IQR_helpful

#Identifying outliers in 'helpful_count':
outliers_helpful = reviews_df[(reviews_df['helpful_count'] < lower_bound_helpful) | (reviews_df['helpful_count'] > upper_bound_helpful)]

print("\nOutliers in 'helpful_count':")
print(outliers_helpful)
```

The dataset includes 184 outliers in the helpful_count column, identified using the Interquartile Range (IQR) method. These outliers represent reviews that received an unusually high number of helpful votes compared to the majority of other reviews. Each outlier entry includes attributes such as review_id, product_id, title, author, rating, content, timestamp, profile_id, is_verified, helpful_count, and product_attributes.

I have chosen not to handle these outliers in my analysis, opting instead to retain them in order to capture the full variability and potential insights they provide.
